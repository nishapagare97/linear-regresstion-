```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_income=pd.read_csv("income.csv")
df_income
```

```python
df_income.plot(x='Year',y='Income',linestyle='dashed',linewidth=2,markersize=15)
plt.title("Income Chart")
plt.xlabel("Year",fontsize=15)
plt.ylabel("Income(IND Rs..)",fontsize=15)
```

```python
#plot the Liner Line On Sctter Grap 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df_income[['Year']],df_income.Income)
plt.scatter(df_income.Year,df_income.Income,color='green',marker="*")
plt.plot(df_income.Year,model.predict(df_income[['Year']]),color="blue")
plt.xlabel("Year",fontsize=15)
plt.ylabel("Income",fontsize=15)
plt.title("Regression line",fontsize=10)
```

```python
#predict income for year 2020
model.predict([[2020]])
```

```python
#predict income for year 2021
model.predict([[2021]])
```
